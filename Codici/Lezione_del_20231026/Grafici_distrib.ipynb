```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
n, p = 5, 0.4
```

```python
x = np.arange(0., n+1.)
```

```python
y = binom.pmf(x, n, p)
```

```python
plt.plot(x, y, '*')
plt.xlabel('numero di successi')
plt.ylabel('probabilit√†')
plt.show()
```

```python
xx = np.linspace(0., 7., 100)
z = binom.cdf(xx, n, p)
```

```python
plt.plot(xx, z)
plt.xlabel('x')
plt.ylabel('funzione di ripartizione F(x)')
plt.show()
```
