```python
# Esercizio intervallo di confidenza per la media con varianza ignota
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
```

```python
X = np.array([19.8, 10.1, 14.9, 7.5, 15.4, 15.4, 15.4, 18.5, 7.9, 12.7, 11.9, 11.4, \
              11.4, 14.1, 17.6, 16.7, 15.8, 19.5, 8.8, 13.6, 11.9, 11.4])
```

```python
me = np.mean(X)
S = np.std(X, ddof=1)
```

```python
display(me)
display(S)
```

```python
# crea box_plot
fig, ax = plt.subplots(1, 1)

ax.boxplot(X)
plt.show()
```

```python
# Grafico di probabilità normale
from scipy.stats import probplot
from scipy.stats import norm
```

```python
fig, ax = plt.subplots(1, 1)
probplot(X, dist=norm, plot=ax)
plt.show()
```

```python
# Si può assumere che la distribuzione della popolazione è approssimativamente normale
```

```python
# Calcoliamo l'intervallo di confidenza al 95% per la media
```

```python
n = X.size
df = n-1
```

```python
alpha = 0.05
t = t.ppf(1.-alpha/2., df)
```

```python
Il = me - S/np.sqrt(n)*t
Ir = me + S/np.sqrt(n)*t
```

```python
display(Il,Ir)
```
