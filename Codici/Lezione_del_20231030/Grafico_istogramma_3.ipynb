```python
# Esempio di istogramma
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
edges = np.array([[0, 0.5],[0.5,1],[1,1.5],[1.5,2]])
```

```python
ampiezza = edges[:,1] - edges[:,0]
```

```python
centers = 0.5*(edges[:,0] + edges[:,1])
```

```python
freq = np.array([10, 8, 6, 3])
```

```python
pp = freq/np.sum(freq)
```

```python
freq_cum = np.cumsum(pp)
freq_cum = np.insert(freq_cum, 0, 0)
```

```python
# crea istogramma
fig, ax = plt.subplots(1, 1)
plt.bar(centers, freq, width=ampiezza)

# set titolo
ax.set_title("Indagine polveri sottili")
  
# set assi
ax.set_xlabel('valore polveri')
ax.set_ylabel('frequenza')

plt.show()
```

```python
x = np.array([0, 0.5, 1, 1.5, 2])
alpha = np.array([0.25, 0.5, 0.75])
```

```python
display(freq_cum)
```

```python
jj = np.zeros(3, dtype=int)
for k in range(3):
    for j in range(1,5):
        if alpha[k]>freq_cum[j-1] and alpha[k]<=freq_cum[j]:
            jj[k] = j
display(jj)    
```

```python
q_alpha = np.zeros(3)
for k in range(3):
    j = jj[k]
    q_alpha[k] = x[j-1] + (alpha[k] - freq_cum[j-1])*(x[j]-x[j-1])/(freq_cum[j]-freq_cum[j-1])
display(q_alpha)
```

```python
# crea grafico
fig, ax = plt.subplots(1, 1)

x = np.array([0, 0.5, 1, 1.5, 2])
plt.plot(x,freq_cum,'*',label='F_i')
plt.plot(x,freq_cum,label='F(x)')
plt.plot(q_alpha,alpha,'o',label='q_alpha')

legend = ax.legend(loc='upper left')

# set titolo
ax.set_title("Indagine polveri sottili")
  
# set assi
ax.set_xlabel('valore polveri')
ax.set_ylabel('funzione di ripartizione empirica')

plt.show()
```
