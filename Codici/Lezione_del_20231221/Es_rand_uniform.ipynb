```python
# Generazione di numeri pseudo-casuali con ditribuzione U([a,b])
```

```python
import numpy as np
```

```python
a = 1.
b = 3.
N = 20000
X = np.random.rand(N)
```

```python
display(X)
```

```python
Y = a + X*(b-a)
```

```python
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(a,b,1000)
from scipy.stats import uniform
yy = uniform.pdf(xx, a, b-a)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(Y, density=True)
plt.plot(xx,yy)
ax.set_title("Distribuzione uniforme in [a,b]")
plt.show()
```

```python
from scipy.stats import probplot
fig, ax = plt.subplots(1, 1)
probplot(Y, dist=uniform, plot=ax)
plt.show()
```
