```python
# Generazione di numeri pseudo-casuali con ditribuzione EXP(lambda)
```

```python
import numpy as np
```

```python
lam = 2.
N = 200000
X = np.random.rand(N)
```

```python
display(X)
```

```python
Y = -np.log(X)/lam
```

```python
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(0.,5.,1000)
from scipy.stats import expon
yy = expon.pdf(xx, 0., 1./lam)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(Y, bins=100, density=True)
plt.plot(xx,yy)
ax.set_title("Distribuzione EXP(lam)")
plt.show()
```

```python
from scipy.stats import probplot
fig, ax = plt.subplots(1, 1)
probplot(Y, dist=expon, plot=ax)
plt.show()
```
