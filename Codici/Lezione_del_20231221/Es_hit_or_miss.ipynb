```python
import numpy as np
```

```python
def fun(x):
    y = np.exp(-x**2./2.)/np.sqrt(2.*np.pi)
    return y
```

```python
a = 0.5
b = 2.
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
xx = np.linspace(-3.,3.,1000)
yy = norm.pdf(xx)
plt.plot(xx,yy)
plt.show()
```

```python
M = 0.45
N = 1000000
NS = 0
for i in range(N):
    r1 = np.random.rand()
    r2 = np.random.rand()
    xi = a + r1*(b-a)
    eta = r2*M
    if fun(xi) > eta:
        NS = NS+1
p = NS/N
I = p*M*(b-a)
display(I)
```

```python
phi = norm.cdf(b)-norm.cdf(a)
display(phi)
```
