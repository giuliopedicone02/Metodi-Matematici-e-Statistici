```python
# Generazione di numeri pseudo-casuali con ditribuzione EXP(lambda)
```

```python
import numpy as np
```

```python
N = 200000
X = np.random.rand(N)
```

```python
display(X)
```

```python
xi1 = X[0:int(N/2)]
xi2 = X[int(N/2):N]
eta1 = np.sqrt(-2*np.log(xi1))*np.cos(2.*np.pi*xi2)
eta2 = np.sqrt(-2*np.log(xi1))*np.sin(2.*np.pi*xi2)
Y = np.zeros(N)
Y[0:int(N/2)] = eta1
Y[int(N/2):N] = eta2
```

```python
import matplotlib.pyplot as plt
```

```python
xx = np.linspace(-5.,5.,1000)
from scipy.stats import norm
yy = norm.pdf(xx)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(Y, bins=100, density=True)
plt.plot(xx,yy)
ax.set_title("Distribuzione N(0,1)")
plt.show()
```

```python
from scipy.stats import probplot
fig, ax = plt.subplots(1, 1)
probplot(Y, dist=norm, plot=ax)
plt.show()
```
