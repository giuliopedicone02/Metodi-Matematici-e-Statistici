```python
# Metodo del rigetto per generare numeri pseudo-casuali
# con distribuzione f(x) = (1+cos(x))/(2*pi)
```

```python
import numpy as np
```

```python
def fun(x):
    y = (1.+np.cos(x))/(2.*np.pi)
    return y
```

```python
def rigetto(a,b,M):
    while True:
        r1 = np.random.rand()
        r2 = np.random.rand()
        xi = a + r1*(b-a)
        eta = M*r2
        if eta <= fun(xi):
            break
    return xi   
```

```python
a = -np.pi
b = np.pi
M = 1./np.pi
```

```python
N = 100000
X = np.zeros(N)
for i in range(N):
    X[i] = rigetto(a,b,M)
```

```python
xx = np.linspace(a,b,1000)
yy = fun(xx)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
ax.hist(X, bins=100, density=True)
plt.plot(xx,yy)
ax.set_title("Distribuzione f(x)")
plt.show()
```
