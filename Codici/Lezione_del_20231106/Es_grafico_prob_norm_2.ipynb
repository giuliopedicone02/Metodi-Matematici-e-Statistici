```python
import numpy as np
import matplotlib.pyplot as plt
# Grafico di probabilità normale
from scipy.stats import probplot
from scipy.stats import norm
```

```python
DATA = np.loadtxt('Data_prob_plot.dat')
X = DATA[:,0]
Y = DATA[:,1]
Z = DATA[:,2]
```

```python
fig, ax = plt.subplots(1, 1)
probplot(X, dist=norm, plot=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
probplot(Y, dist=norm, plot=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
probplot(Z, dist=norm, plot=ax)
plt.show()
```
