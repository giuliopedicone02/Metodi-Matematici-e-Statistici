```python
# Esempio di istogramma
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.loadtxt('Data_altezze.dat')
```

```python
display(X)
```

```python
# Calcolare media e deviazione standard della popolazione
```

```python
media = np.mean(X)
dev_std = np.std(X)
```

```python
display(media)
```

```python
display(dev_std)
```

```python
# Costruire un istogramma a 20 barre

fig, ax = plt.subplots(1, 1)
# ax.hist(X)
ax.hist(X, bins=20, edgecolor = "black")

ax.set_title("Indagine sull'altezza di una popolazione")
ax.set_xlabel('altezza (cm)')
ax.set_ylabel('frequenze')

plt.show()
```

```python
# E' possibile adattare i dati ad una distribuzione normale?
```

```python
from scipy.stats import skew
from scipy.stats import kurtosis
```

```python
s = skew(X)
display(s)
```

```python
k = kurtosis(X, fisher=False) # si veda la documentazione
display(k)
```

```python
from scipy.stats import norm
xx = np.linspace(130, 210, num=200)
yy = norm.pdf(xx, loc=media, scale=dev_std)
```

```python
fig, ax = plt.subplots(1, 1)
# ax.hist(X)
ax.hist(X, density=True, bins=40, edgecolor = "black")

ax.set_title("Indagine sull'altezza di una popolazione")
ax.set_xlabel('altezza (cm)')
ax.set_ylabel('frequenze')

plt.plot(xx,yy)
plt.show()
```
